3. 이미지 처리를 위해 opencv 와 노트북에 이미지를 출력하기 위한 matplotlib를 읽어온다.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
print("읽어오기 완료")
```

4. openCV로 준비한 이미지를 읽는다.
:이미지의 경로를 명시해준다. 

```python
import os
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/image.jpg'
img_bgr = cv2.imread(my_image_path)    #- OpenCV로 이미지를 읽어서
img_bgr = cv2.resize(img_bgr, (500, 650))    # 500x650의 크기로 Resize
img_show = img_bgr.copy()      #- 출력용 이미지 별도 보관

plt.imshow(img_bgr)
plt.show()

# opencv는 RGB가 아닌 BGR를 사용해서 푸른빛이 감돌아보인다. 
# 나중에 다시 바꾸어주어야한다. 
```

5. dlib을 활용해 hog detector를 선언

- 먼저 Object detection 기술을 이용해서 얼굴의 위치를 찾아야한다.
- 그래디언트를 특징으로 사용한다, 
- 지금은 학습 없이 공개되어 있는 패키지를 이용할 것이다.
- dlib 의 face detector는 HOG(Histogram of Oriented Gradient) feature를 사용해서 SVM(Support Vector Machine)의 sliding window로 얼굴을 찾는다. 

```python
import dlib
detector_hog = dlib.get_frontal_face_detector()   #- detector 선언
print("hog detector 선언 완료")
```

6. detector를 이용해서 얼굴의 bounding box를 추출한다. 
- dlib은 rgb 이미지를 입력으로 받기 때문에 cvtColor() 를 이용해서 opencv 의 bgr 이미지를 rgb로 변환한다. 

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) # rgb로 색 변환
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
print("얼굴 bounding box 추출")
```

7. 찾은 얼굴영역을 출력한다.

```python
print(dlib_rects)   # 찾은 얼굴영역 좌표

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

- dlib detector 는 dlib.rectangles 타입의 객체를 반환한다. 
- dlib.rectangles 는 dlib.rectangle 객체의 배열 형태로 이루어져 있다. 
- dlib.rectangle객체는 left(), top(), right(), bottom(), height(), width() 등의 멤버 함수를 포함한다.

# 얼굴 랜드마크 Face Landmark

8. 스티커를 섬세하게 적용하기 위해서는 이목구비의 위치를 아는 것이 중요하다. landmark를 찾기 위해 Dlib에서 제공되는 학습모델 다운받는다. (없을 경우 다운)

$ cd ~/aiffel/camera_sticker && bzip2 -d ./models/shape_predictor_68_face_landmarks.dat.bz2

9. 저장한 landmark모델을 불러온다. 
- landmark_predictor 는 RGB 이미지와 dlib.rectangle을 입력 받아 dlib.full_object_detection 를 반환한다. 

```python
import os
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("landmark모델 불러오기 완료")
```

__landmark_predictor__ 는 RGB 이미지와 dlib.rectangle을 입력 받아 dlib.full_object_detection 를 반환한다. 

10. points 는 dlib.full_object_detection 의 객체이기 때문에 parts() 함수로 개별 위치에 접근할 수 있습니다. 조금 더 직관적인 (x, y) 형태로 접근할 수 있도록 변환해 주었습니다.따라서 list_points 는 tuple (x, y) __68개__ 로 이루어진 리스트가 된다. 

```python
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

11. 랜드마크를 영상에 출력한다. 

```python
for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) # yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

# 스티커 적용하기

12. 코의 좌표를 확인하고 스티커가 들어갈 위치를 계산한다.. 

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print ("코의 중심점은", landmark[8]) # nose center index : 30
    x = landmark[30][0] 
    y = landmark[30][1]
    w = dlib_rect.width()
    h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

13. 준비해 둔 스티커 이미지를 읽어서 적용한다. 

```python
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

14. 원본 이미지에 스티커 이미지를 추가하기 위해서 x, y 좌표를 조정한다. 
(이미지 시작점은 top-left 좌표이기 때문)

```python
refined_x = x - w//2     # left
refined_y = y - h//2    # top
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

15. 만약 스티커가 위로 삐져나온다면 자른다. 

```python
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:]
    print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
    refined_y = 0
```

16. 원본 이미지에 스티커를 적용한다. 

```python
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]= \
	np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

```python
#결과출력
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]= \
	np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
print("고양이콧수염!")
```

# 콧수염 불투명도 설정

```python
sticker_area = img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]= \
	cv2.addWeighted(img_sticker,0.3,sticker_area ,0.7, 0)
```

```python
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
